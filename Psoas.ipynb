```python
import nibabel as nib
import numpy as np
from skimage.transform import resize

def read_nii(filepath):
    '''
    Reads a '.nii' file and returns a numpy array
    '''
    nii_data = nib.load(filepath)
    return nii_data.get_fdata()

def preprocess_data(ct_array, mask_array, img_depth=196, new_shape=(128, 128)):
    '''
    Preprocess the data: normalize CT images, resize images and masks.
    Assumes CT and mask arrays are [height, width, depth].
    '''
    # Normalize CT images
    ct_array_norm = ct_array / np.max(ct_array)

    # Resize images and masks
    resized_ct = np.zeros((img_depth, new_shape[0], new_shape[1]))
    resized_mask = np.zeros((img_depth, new_shape[0], new_shape[1]))
    for i in range(img_depth):
        #print("çt = ", ct_array_norm[:, :, i].shape)
        resized_ct[i, :, :] = resize(ct_array_norm[:, :, i], new_shape, mode='constant', anti_aliasing=True)
        #print(resized_ct[i, :, :].shape)
        resized_mask[i, :, :] = resize(mask_array[:, :, i], new_shape, mode='constant', anti_aliasing=True)

    # Add a channel dimension
    resized_ct = resized_ct[..., np.newaxis]
    resized_mask = resized_mask[..., np.newaxis]

    return resized_ct, resized_mask

# Load the data
ct_2_mask = read_nii("file_2_mask.nii")
ct_2_main = read_nii("file_2_ct.nii")

# Assuming these arrays are shaped as [height, width, depth]
preprocessed_ct, preprocessed_mask = preprocess_data(ct_2_main, ct_2_mask)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate

def unet_model(input_size=(128, 128, 1)):
    inputs = Input(input_size)
    # Downsample
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(pool1)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    # Bottleneck
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(pool2)

    # Upsample
    up1 = UpSampling2D(size=(2, 2))(conv3)
    merge1 = concatenate([conv2, up1], axis=3)
    conv4 = Conv2D(128, 3, activation='relu', padding='same')(merge1)
    up2 = UpSampling2D(size=(2, 2))(conv4)
    merge2 = concatenate([conv1, up2], axis=3)
    conv5 = Conv2D(64, 3, activation='relu', padding='same')(merge2)

    # Output layer
    conv6 = Conv2D(1, 1, activation='sigmoid')(conv5)

    model = Model(inputs=inputs, outputs=conv6)

    return model

# Initialize U-Net model
unet = unet_model()
```

```python
from sklearn.model_selection import train_test_split

# Assuming you have more data and have followed the preprocessing step for each
X_train, X_val, y_train, y_val = train_test_split(preprocessed_ct, preprocessed_mask, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras import backend as K
from sklearn.metrics import confusion_matrix

def dice_coef(y_true, y_pred, smooth=1):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def iou(y_true, y_pred, smooth=1):
    intersection = K.sum(K.abs(y_true * y_pred), axis=-1)
    sum_ = K.sum(K.abs(y_true) + K.abs(y_pred), axis=-1)
    jac = (intersection + smooth) / (sum_ - intersection + smooth)
    return jac
```

```python
import nibabel as nib
import numpy as np
from skimage.transform import resize


def combine_volumes(paths):
    '''
    Read, preprocess, and combine volumes from a list of file paths
    '''
    all_volumes = []
    for path in paths:
        image = read_nii(path)

        all_volumes.append(image)
    # Combine along depth axis
    combined_volume = np.concatenate(all_volumes, axis=2)
    return combined_volume


# I add the threshold to make sure our mask only contains zeros and ones
def preprocess_data(ct_array, mask_array, mask_threshold=0.5, new_shape=(128, 128)):
    '''
    Preprocess the data: normalize CT images, resize images and masks.
    Assumes CT and mask arrays are [height, width, depth].
    '''
    # Normalize CT images
    ct_array_norm = ct_array / np.max(ct_array)

    img_depth = ct_array.shape[2]

    # Resize images and masks
    resized_ct = np.zeros((img_depth, new_shape[0], new_shape[1]))
    resized_mask = np.zeros((img_depth, new_shape[0], new_shape[1]))
    for i in range(img_depth):

        resized_ct[i, :, :] = resize(ct_array_norm[:, :, i], new_shape, mode='constant', anti_aliasing=True)
        resized_mask[i, :, :] = resize(mask_array[:, :, i], new_shape, mode='constant', anti_aliasing=True)

        resized_mask[i, :, :] = np.where(resized_mask[i, :, :] < mask_threshold, 0, 1)
    # Add a channel dimension
    resized_ct = resized_ct[..., np.newaxis]
    resized_mask = resized_mask[..., np.newaxis]

    return resized_ct, resized_mask


ct_image_paths = ["file_ct_0029.nii", "file_ct_0166.nii"]  # List of CT image file paths
mask_paths = ["file_mask_0029.nii", "file_mask_0166.nii"]  # List of mask file paths

# Assuming each volume to be resized to 128x128 per slice and each set of volumes to have 196 slices
new_shape = (128, 128)

# Combine CT images and masks separately
combined_cts = combine_volumes(ct_image_paths) # , img_depth=img_depth_per_volume
combined_masks = combine_volumes(mask_paths) # , img_depth=img_depth_per_volume

combined_cts.shape
```

```python
import tensorflow as tf
import sys


# Custom accuracy metric function that incorporates class weights
def weighted_accuracy(y_true, y_pred):

    # Apply argmax to get predicted class labels
    y_pred_labels = tf.argmax(y_pred, axis=-1, output_type=tf.int64)

    print('y_pred_labels.shape = ', y_pred_labels.shape)
    print('y_true.shape = ', y_true.shape)
    print('y_pred_labels.dtype = ', y_pred_labels.dtype)
    # tf.print(y_true, output_stream=sys.stderr, sep=',')

    y_true = tf.cast(y_true, dtype=y_pred_labels.dtype)
    print('y_true.shape = ', y_true.shape)

    # Compute correct predictions
    correct_predictions = tf.cast(tf.equal(x=y_true, y=y_pred_labels), dtype=tf.float32)
    print('after correct_predictions.shape =  ', correct_predictions.shape)

    # Multiply correct predictions by class weights
    weighted_correct_predictions = correct_predictions * class_weights_dict[y_true]
    print('after weighted_correct_predictions.shape =  ', weighted_correct_predictions.shape)

    # Compute total weighted correct predictions
    total_weighted_correct_predictions = tf.reduce_sum(weighted_correct_predictions)
    print('after total_weighted_correct_predictions.shape =  ', total_weighted_correct_predictions.shape)

    # Compute total weight of all samples
    total_weight = tf.reduce_sum(class_weights_dict[y_true])

    # Compute weighted accuracy
    weighted_accuracy = total_weighted_correct_predictions / total_weight

    return weighted_accuracy
```

```python

# Visualize a few predictions
def sample_predictions(X_test, num_samples_to_visualize = 3):

    """
    visualizing a sample of num_samples_to_visualize (e.g., 3) images
    for depicting the prediction progress during the training.
    """

    # number of samples found with 1 in their prediction pixels
    aquired_n = 0

    # randomly choose 50 and check each for masks
    indices_to_visualize = np.random.choice(len(X_test), 50, replace=False)
    idx_vis = []

     # go through them until we get that many that we want.
    while aquired_n != num_samples_to_visualize:
        indices_to_visualize = np.random.choice(len(X_test), 50, replace=False)

        for idx in indices_to_visualize:
            if np.sum(X_test[idx]) != 0 and aquired_n != num_samples_to_visualize:
                #print(f'found one ! -> {idx}')
                idx_vis.append(idx)
                aquired_n += 1
    return idx_vis
```

```python
import onedrivesdk

redirect_uri = 'http://localhost:8080/' client_secret = 'your_client_secret' client_id='your_client_id' api_base_url='https://api.onedrive.com/v1.0/'
scopes=['wl.signin', 'wl.offline_access', 'onedrive.readwrite']
http_provider = onedrivesdk.HttpProvider()
auth_provider = onedrivesdk.AuthProvider( http_provider=http_provider, client_id=client_id, scopes=scopes)
client = onedrivesdk.OneDriveClient(api_base_url, auth_provider, http_provider)
auth_url = client.auth_provider.get_auth_url(redirect_uri)

# Ask for the code
print('Paste this URL into your browser, approve the app\'s access.')
print('Copy everything in the address bar after "code=", and paste it below.') print(auth_url)
code = input('Paste code here: ')  client.auth_provider.authenticate(code, redirect_uri, client_secret)
```

```python

# Step 2: Import necessary libraries
from pyodrive import ODAuth, ODItem

# Step 3: Authenticate with your OneDrive account
# You'll need to replace 'YOUR_CLIENT_ID' and 'YOUR_CLIENT_SECRET' with your actual client ID and secret obtained from registering your app in Azure portal.
od_auth = ODAuth(client_id='YOUR_CLIENT_ID', client_secret='YOUR_CLIENT_SECRET')

# Authenticate
od_auth.authenticate()

# Step 4: Get the root folder
root_folder = ODItem.get_root(od_auth)

# Step 5: List files in the root folder
for item in root_folder.children:
    print(item.name)

# Step 6: Download files
# Replace 'file_name_to_download' with the name of the file you want to download
file_to_download = root_folder.get_child_by_name('file_name_to_download')

# Download the file
file_to_download.download('path_to_save')

# Replace 'path_to_save' with the path where you want to save the file

# Once you run this code, it will prompt you to authenticate your OneDrive account in Colab. Follow the instructions and provide the necessary permissions. After authentication, you'll be able to list files and download them from your OneDrive.
```
